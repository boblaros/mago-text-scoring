```python
import json
import math
import os
import random
import re
from datetime import datetime
from pathlib import Path

import numpy as np
import pandas as pd
import torch
from IPython.display import display
from sklearn.metrics import accuracy_score, classification_report, f1_score
from torch.utils.data import Dataset
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    EarlyStoppingCallback,
    Trainer,
    TrainingArguments,
)
from transformers.trainer_utils import get_last_checkpoint

os.environ.setdefault("TOKENIZERS_PARALLELISM", "false")
print("Imports OK")
```

```python
CFG = {
    "seed": 42,
    "data_dir": DRIVE_ROOT / "data",
    "output_dir": DRIVE_ROOT / "outputs",
    "model_dir": DRIVE_ROOT / "models",
    "log_dir": DRIVE_ROOT / "logs",
    "task": "complexity-classification",
    "text_col": "text",
    "label_col": "label",
    "encoded_label_col": "label_enc",
    "num_labels": 3,
    "train_pool_key": "train_pool",
    "train_parquet_path": DRIVE_ROOT / "data" / "train_pool_train.parquet",
    "val_parquet_path": DRIVE_ROOT / "data" / "train_pool_val.parquet",
    "test_parquet_path": DRIVE_ROOT / "data" / "train_pool_test.parquet",
    "experiment_dataset_slug": "complexity",
    "max_len": 256,
    "batch_size": 8,
    "epochs": 3,
    "lr": 2e-5,
    "device": "cuda" if torch.cuda.is_available() else "cpu",
}

for key in ["data_dir", "output_dir", "model_dir", "log_dir"]:
    Path(CFG[key]).mkdir(parents=True, exist_ok=True)

RESULTS = {}
CLASSES = ["A1-A2", "B1-B2", "C1-C2"]
class2id = {c: i for i, c in enumerate(CLASSES)}
id2class = {i: c for c, i in class2id.items()}


def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)


class DictLabelEncoder:
    def __init__(self, mapping: dict):
        first_key = next(iter(mapping))
        if isinstance(first_key, int):
            id2class_local = mapping
        else:
            id2class_local = {v: k for k, v in mapping.items()}
        self.classes_ = np.array([id2class_local[i] for i in sorted(id2class_local)])

    def inverse_transform(self, y):
        y_arr = np.asarray(y, dtype=np.int64)
        return np.array([self.classes_[i] if 0 <= i < len(self.classes_) else str(i) for i in y_arr])


def slugify(text: str) -> str:
    text = str(text).strip().lower()
    text = re.sub(r"[^a-z0-9]+", "_", text)
    return re.sub(r"_+", "_", text).strip("_")


def now_ts() -> str:
    return datetime.now().strftime("%Y%m%d_%H%M%S")


def size_tag_from_split(split_key: str) -> str:
    m = re.search(r"(\d+[km]?)$", split_key.lower())
    return m.group(1) if m else slugify(split_key)


def evaluate_clf(model_name: str, y_true, y_pred, label_encoder=None, split: str = "test") -> dict:
    acc = accuracy_score(y_true, y_pred)
    f1_macro = f1_score(y_true, y_pred, average="macro", zero_division=0)
    f1_weight = f1_score(y_true, y_pred, average="weighted", zero_division=0)

    metrics = {
        "accuracy": float(acc),
        "f1_macro": float(f1_macro),
        "f1_weighted": float(f1_weight),
    }
    RESULTS[f"{model_name} | {split}"] = metrics

    print(f"\n{model_name} | {split.upper()}")
    print(metrics)
    if label_encoder is not None:
        print(classification_report(
            y_true,
            y_pred,
            labels=np.arange(len(label_encoder.classes_)),
            target_names=list(label_encoder.classes_),
            zero_division=0,
        ))
    else:
        print(classification_report(y_true, y_pred, zero_division=0))

    return metrics


def find_latest_experiment_dir(prefix: str, model_root: Path = None):
    model_root = Path(model_root or CFG["model_dir"])
    model_root.mkdir(parents=True, exist_ok=True)
    patt = re.compile(rf"^{re.escape(prefix)}_\d{{8}}_\d{{6}}$")
    candidates = [p for p in model_root.iterdir() if p.is_dir() and patt.match(p.name)]
    if not candidates:
        return None
    return sorted(candidates, key=lambda p: p.name)[-1]


def resolve_experiment_paths(model_family: str, dataset: str, size_tag: str, model_root: Path = None) -> dict:
    model_root = Path(model_root or CFG["model_dir"])
    model_root.mkdir(parents=True, exist_ok=True)
    prefix = f"{slugify(model_family)}_{slugify(dataset)}_{slugify(size_tag)}"
    exp_root = find_latest_experiment_dir(prefix, model_root=model_root)
    if exp_root is None:
        exp_root = model_root / f"{prefix}_{now_ts()}"

    paths = {
        "prefix": prefix,
        "root": exp_root,
        "best_model": exp_root / "best_model",
        "checkpoints": exp_root / "checkpoints",
        "runs": exp_root / "runs",
    }
    for p in paths.values():
        if isinstance(p, Path):
            p.mkdir(parents=True, exist_ok=True)
    return paths


class HFTextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len: int = 256):
        self.texts = np.asarray(texts, dtype=object)
        self.labels = np.asarray(labels, dtype=np.int64)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            str(self.texts[idx]),
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt",
        )
        item = {k: v.squeeze(0) for k, v in enc.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item


def compute_metrics(eval_pred) -> dict:
    logits, labels = eval_pred
    if isinstance(logits, tuple):
        logits = logits[0]
    preds = np.argmax(logits, axis=1)
    return {
        "f1_macro": f1_score(labels, preds, average="macro", zero_division=0),
        "accuracy": accuracy_score(labels, preds),
    }


def resolve_text_col(df: pd.DataFrame) -> str:
    for col in ["text_raw", CFG["text_col"], "text_clean"]:
        if col in df.columns:
            return col
    raise ValueError(f"Missing text column in {sorted(df.columns)}")


def resolve_target_col(df: pd.DataFrame) -> str:
    for col in [CFG["encoded_label_col"], "label_enc"]:
        if col in df.columns:
            return col
    raise ValueError(f"Missing encoded label column in {sorted(df.columns)}")


def load_split_parquet(data_path: Path, split_name: str) -> pd.DataFrame:
    data_path = Path(data_path)
    if not data_path.exists():
        raise FileNotFoundError(
            f"Missing parquet for {split_name}: {data_path}\n"
            "Upload the exported train_pool parquet files to DRIVE_ROOT/data before running this notebook."
        )

    df = pd.read_parquet(data_path).reset_index(drop=True)
    text_col = resolve_text_col(df)
    target_col = resolve_target_col(df)

    frame = df.copy()
    frame["text_raw"] = frame[text_col].astype(str)
    frame[CFG["encoded_label_col"]] = pd.to_numeric(frame[target_col], errors="raise").astype("int64")
    if CFG["label_col"] not in frame.columns:
        frame[CFG["label_col"]] = frame[CFG["encoded_label_col"]].map(id2class)
    return frame


def build_transformer_bundle_from_parquets(train_path: Path, val_path: Path, test_path: Path) -> dict:
    df_train = load_split_parquet(train_path, "train")
    df_val = load_split_parquet(val_path, "val")
    df_test = load_split_parquet(test_path, "test")

    return {
        "name": CFG["train_pool_key"],
        "df_train": df_train.copy(),
        "df_val": df_val.copy(),
        "df_test": df_test.copy(),
        "X_train_raw": np.asarray(df_train["text_raw"], dtype=object),
        "X_val_raw": np.asarray(df_val["text_raw"], dtype=object),
        "X_test_raw": np.asarray(df_test["text_raw"], dtype=object),
        "y_train": np.asarray(df_train[CFG["encoded_label_col"]], dtype=np.int64),
        "y_val": np.asarray(df_val[CFG["encoded_label_col"]], dtype=np.int64),
        "y_test": np.asarray(df_test[CFG["encoded_label_col"]], dtype=np.int64),
    }


def split_summary_row(split_name: str, df: pd.DataFrame, path: Path) -> dict:
    counts = df[CFG["encoded_label_col"]].value_counts().sort_index().to_dict()
    named_counts = {id2class.get(int(k), str(k)): int(v) for k, v in counts.items()}
    return {
        "split": split_name,
        "rows": int(len(df)),
        "labels": json.dumps(named_counts, ensure_ascii=False),
        "path": str(path),
    }


def _build_training_args(common_args: dict, eval_mode: str = "epoch"):
    args = dict(common_args)
    args["save_strategy"] = eval_mode
    if "eval_strategy" in TrainingArguments.__dataclass_fields__:
        args["eval_strategy"] = eval_mode
    else:
        args["evaluation_strategy"] = eval_mode
    return TrainingArguments(**args)


def run_transformer_experiment(
    display_name: str,
    family_slug: str,
    size_tag: str,
    model_name: str,
    data_bundle: dict,
    batch_size: int,
    epochs: int,
    lr: float,
    warmup_ratio: float = 0.06,
    weight_decay: float = 0.01,
    gradient_accumulation_steps: int = 1,
    early_stopping_patience: int = 2,
):
    exp_paths = resolve_experiment_paths(
        model_family=family_slug,
        dataset=CFG["experiment_dataset_slug"],
        size_tag=size_tag,
        model_root=CFG["model_dir"],
    )

    best_model_dir = exp_paths["best_model"]
    has_best_model = (best_model_dir / "config.json").exists()

    tokenizer_src = str(best_model_dir) if has_best_model else model_name
    trf_tokenizer = AutoTokenizer.from_pretrained(tokenizer_src, use_fast=True)

    trf_ds_train = HFTextDataset(data_bundle["X_train_raw"], data_bundle["y_train"], trf_tokenizer, CFG["max_len"])
    trf_ds_val = HFTextDataset(data_bundle["X_val_raw"], data_bundle["y_val"], trf_tokenizer, CFG["max_len"])
    trf_ds_test = HFTextDataset(data_bundle["X_test_raw"], data_bundle["y_test"], trf_tokenizer, CFG["max_len"])

    label_names = list(getattr(le, "classes_", CLASSES))
    id2label_local = {i: label_names[i] for i in range(len(label_names))}
    label2id_local = {label: i for i, label in id2label_local.items()}

    if not has_best_model:
        model = AutoModelForSequenceClassification.from_pretrained(
            model_name,
            num_labels=CFG["num_labels"],
            id2label=id2label_local,
            label2id=label2id_local,
            problem_type="single_label_classification",
        )

        effective_bs = max(1, batch_size * gradient_accumulation_steps)
        steps_per_epoch = max(1, math.ceil(len(trf_ds_train) / effective_bs))
        warmup_steps = max(1, int(steps_per_epoch * epochs * warmup_ratio))

        common_args = dict(
            output_dir=str(exp_paths["checkpoints"]),
            num_train_epochs=epochs,
            per_device_train_batch_size=batch_size,
            per_device_eval_batch_size=batch_size,
            gradient_accumulation_steps=gradient_accumulation_steps,
            learning_rate=lr,
            warmup_steps=warmup_steps,
            weight_decay=weight_decay,
            save_total_limit=2,
            load_best_model_at_end=True,
            metric_for_best_model="f1_macro",
            greater_is_better=True,
            logging_strategy="steps",
            logging_steps=100,
            report_to="none",
            dataloader_num_workers=0,
            fp16=(CFG["device"] == "cuda"),
            bf16=False,
            seed=CFG["seed"],
        )

        trainer = Trainer(
            model=model,
            args=_build_training_args(common_args, eval_mode="epoch"),
            train_dataset=trf_ds_train,
            eval_dataset=trf_ds_val,
            compute_metrics=compute_metrics,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=early_stopping_patience)],
        )

        last_ckpt = get_last_checkpoint(str(exp_paths["checkpoints"]))
        print(f"{display_name}: last checkpoint -> {last_ckpt}")
        trainer.train(resume_from_checkpoint=last_ckpt if last_ckpt is not None else None)

        trainer.save_model(str(best_model_dir))
        trf_tokenizer.save_pretrained(str(best_model_dir))
        print(f"{display_name}: training complete -> {best_model_dir}")
    else:
        print(f"{display_name}: best model already exists -> skip retraining ({best_model_dir})")

    eval_model = AutoModelForSequenceClassification.from_pretrained(str(best_model_dir))
    eval_args = {
        "output_dir": str(exp_paths["runs"] / "eval_tmp"),
        "per_device_eval_batch_size": batch_size,
        "dataloader_num_workers": 0,
        "report_to": "none",
        "seed": CFG["seed"],
        "fp16": False,
        "bf16": False,
    }
    if "eval_strategy" in TrainingArguments.__dataclass_fields__:
        eval_args["eval_strategy"] = "no"
    else:
        eval_args["evaluation_strategy"] = "no"

    eval_trainer = Trainer(
        model=eval_model,
        args=TrainingArguments(**eval_args),
        compute_metrics=compute_metrics,
    )

    val_output = eval_trainer.predict(trf_ds_val)
    test_output = eval_trainer.predict(trf_ds_test)

    val_preds = np.argmax(val_output.predictions, axis=1)
    test_preds = np.argmax(test_output.predictions, axis=1)

    val_metrics = evaluate_clf(display_name, data_bundle["y_val"], val_preds, label_encoder=le, split="val")
    test_metrics = evaluate_clf(display_name, data_bundle["y_test"], test_preds, label_encoder=le, split="test")

    run_metrics = {
        "val": val_metrics,
        "test": test_metrics,
        "best_model_dir": str(best_model_dir),
        "checkpoints_dir": str(exp_paths["checkpoints"]),
        "experiment_dir": str(exp_paths["root"]),
    }

    with open(exp_paths["runs"] / "metrics_latest.json", "w", encoding="utf-8") as f:
        json.dump(run_metrics, f, indent=2)

    np.save(exp_paths["runs"] / "val_preds_latest.npy", val_preds)
    np.save(exp_paths["runs"] / "test_preds_latest.npy", test_preds)

    return {
        "name": display_name,
        "val_preds": val_preds,
        "test_preds": test_preds,
        "y_val": data_bundle["y_val"],
        "y_test": data_bundle["y_test"],
        "df_test": data_bundle["df_test"].copy().reset_index(drop=True),
        "best_dir": best_model_dir,
        "paths": exp_paths,
        "val_metrics": val_metrics,
        "test_metrics": test_metrics,
    }


set_seed(CFG["seed"])
le = DictLabelEncoder(class2id)
print(f"Device: {CFG['device']}")
```

```python
BASE_TRAIN_POOL_BUNDLE = build_transformer_bundle_from_parquets(
    CFG["train_parquet_path"],
    CFG["val_parquet_path"],
    CFG["test_parquet_path"],
)
TRANSFORMER_SIZE_TAG = size_tag_from_split(CFG["train_pool_key"])

summary_df = pd.DataFrame([
    split_summary_row("train", BASE_TRAIN_POOL_BUNDLE["df_train"], CFG["train_parquet_path"]),
    split_summary_row("val", BASE_TRAIN_POOL_BUNDLE["df_val"], CFG["val_parquet_path"]),
    split_summary_row("test", BASE_TRAIN_POOL_BUNDLE["df_test"], CFG["test_parquet_path"]),
])
display(summary_df)
```

```python
print("### DeBERTa-v3-base ###")

DEBERTA_RUN = run_transformer_experiment(
    display_name="DeBERTaV3_train_pool",
    family_slug="deberta_v3_base",
    size_tag=TRANSFORMER_SIZE_TAG,
    model_name="microsoft/deberta-v3-base",
    data_bundle=BASE_TRAIN_POOL_BUNDLE,
    batch_size=min(CFG["batch_size"], 8),
    epochs=CFG["epochs"],
    lr=2e-5,
    warmup_ratio=0.06,
    weight_decay=0.01,
    gradient_accumulation_steps=2,
    early_stopping_patience=2,
)
```

```python
deberta_summary_df = pd.DataFrame([
    {
        "experiment": DEBERTA_RUN["name"],
        "val_f1_macro": float(DEBERTA_RUN["val_metrics"]["f1_macro"]),
        "test_f1_macro": float(DEBERTA_RUN["test_metrics"]["f1_macro"]),
        "val_accuracy": float(DEBERTA_RUN["val_metrics"]["accuracy"]),
        "test_accuracy": float(DEBERTA_RUN["test_metrics"]["accuracy"]),
        "best_model": str(DEBERTA_RUN["best_dir"]),
    }
])

display(deberta_summary_df)

summary_path = DEBERTA_RUN["paths"]["runs"] / "deberta_summary_latest.csv"
deberta_summary_df.to_csv(summary_path, index=False)
print(f"Saved summary to: {summary_path}")
print(f"Best model dir: {DEBERTA_RUN['best_dir']}")
```
